## We get all the test matches playes between two nations since 2022

```python
import pandas as pd
```

#### Loading data from csv file

```python
cric_data = pd.read_csv("/content/cricket_test_matches.csv")
cric_data.head()
```

## Function to get the matches

```python
def get_matches_between_teams(team1, team2, data):
    filtered_matches = data[((data['Team1'] == team1) & (data['Team2'] == team2)) |
                            ((data['Team1'] == team2) & (data['Team2'] == team1))]
    return filtered_matches
```

```python
team1 = input("Enter the first team: ")
team2 = input("Enter the second team: ")

matches = get_matches_between_teams(team1, team2, cric_data)
```

```python
matches
```

```python
matches.to_csv("/content/matches_between_{}_and_{}.csv".format(team1, team2), index=False)

print(f"Filtered matches between {team1} and {team2} have been saved to 'matches_between_{team1}_and_{team2}.csv'.")
```
